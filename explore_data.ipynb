# Data Exploration for Hackathon 2026
This notebook provides a basic overview of the provided SQLite database.

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('DB/db.sqlite')

# List all tables
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tables in the database:")
print(tables)
```

```python
# Explore the Company table
company_df = pd.read_sql_query("SELECT * FROM Company LIMIT 5;", conn)
print("Company table (first 5 rows):")
company_df
```

```python
# Explore the BOM table
bom_df = pd.read_sql_query("SELECT * FROM BOM LIMIT 5;", conn)
print("BOM table (first 5 rows):")
bom_df
```
